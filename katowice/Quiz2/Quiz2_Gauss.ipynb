## Plot sample of  gaussian variables

If we approximate our set with the Gaussian distribution, which is the product of:

$$P(x,y)  = P(x) P(y)$$

then we can draw a sample easily using a Gaussian random number generator with mean zero and one variance. Note that you should scale such numbers yourself, multiply them before the standard deviation and adding the average value.

Compare 1000 data so generated with real data from the first class.

```python
plt.plot(X[:,0],X[:,1],'.')
plt.plot(m1+s1*np.random.randn(1000),m2+s2*np.random.randn(1000),'.')
```
