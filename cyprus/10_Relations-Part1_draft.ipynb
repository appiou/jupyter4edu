## Definition of Relation

* Any subset of $A \times B$ is called a <font color='red'> **relation**</font>, which is denoted by $R$. 
In other words, we have that a relation $R \subset A \times B$ and we sometimes write this as 
    $ a\ R\ b$ (read as: $a$ is related to $b$)
    
* A relation $R$ can be treated as a set of ordered pairs:
$$R= \{ (a,b) \in A \times B\ |\; a\ R\ b \}$$

    
* Have in mind that $ a\ R\ b$  actually means that the pair $(a,b) \in R$


* Note that if $A=B$, then $R$ is called a **binary** relation.

### Some comments
* Given the sets $A$ and $B$, a relation on $A \times B$ is some property that is either **true** or **false** for any ordered pair $(a, b) \in A \times B$.

* The concept of a relation is as general as the concept of a set, and is not limited to strictly mathematical settings.

* Relations are more general than functions. Recall that a function is a relation where exactly one element of $B$ is related to each element of $A$.

## Binary Relations

Given a set $A$, a <font color='red'> ** binary relation**</font> $R$ on $A$ is a subset of $A \times A$ (i.e. $R \subset A \times A$)

### Example:
If $A = \{1, 2, 3\}$. 
Then $A \times A =$? 

Now, define the relation $R$ on $A = \{1, 2, 3\}$ to be given by 
$$a\ R\ b \leftrightarrow a+b\ \mbox{is odd.}$$ 
Then  $R =$? 

**Question**: How many relations are there on a set $A$ with $n$ elements? 

**Answer**:  Because a relation on $A$ is the same thing as a subset of $A \times A$, we count the subsets of $A \times A$. Since $A \times A$ has $n \times n = n^2$ elements, and a set with $m$ elements has $2^m$ subsets, there are $2^{n^2}$ subsets of $A \times A$. Therefore, there are $2^{n^2}$ relations on a set $A$.


## Various types of Binary Relations

**One-to-one relation**: each first component and each second component appear only once in the relation. 

**One-to-many relation**: if some first component appears more than once.

**Many-to-one relation**: if some second component is paired with more than one first component. 

**Many-to-many relation**: if at least one first component is paired with more than one second component and at least one second is paired with more than one first component.

## Composition of Relations

* Consider the following
    * Arbitrary, finite sets $ A, B, C $
    * A relation from $A$ to $B$, being $R_1$
    * A relation from $B$ to $C$, being $R_2$
* This means that $R_1$ is a subset of $ A \times B $ and $R_2$ is a subset of $ B \times C $
* Now there is a relation for $A$ to $C$, namely $ R_2 \circ R_1 $
    * $ a \left( R_2\circ R_1 \right) c  $ if for some $ b \in B $ we have $ a\ R_1\ b $ and $ b\ R_2\ c $
* This relation is called the <font color='red'> **composition** </font> of relations $R_1$ and $R_2$

***
