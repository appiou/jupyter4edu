```python
t=np.linspace(0.,100.,500)
x=np.sin(2*np.pi*t)
y=np.cos(2*np.pi*t)

# Mercury
xMercury=0.387*np.sin(2*np.pi*t*0.241)
yMercury=0.387*np.cos(2*np.pi*t*0.241)

# Venus
xVenus=0.723*np.sin(2*np.pi*t*0.615)
yVenus=0.723*np.cos(2*np.pi*t*0.615)

# Mars
xMars=1.52*np.sin(2*np.pi*t*1.88)
yMars=1.52*np.cos(2*np.pi*t*1.88)

# Jupiter
xJupiter=5.2*np.sin(2*np.pi*t*11.9)
yJupiter=5.2*np.cos(2*np.pi*t*11.9)

# Saturn
xSaturn=9.58*np.sin(2*np.pi*t*29.4)
ySaturn=9.58*np.cos(2*np.pi*t*29.4)

# Uranus
xUranus=19.2*np.sin(2*np.pi*t*83.7)
yUranus=19.2*np.cos(2*np.pi*t*83.7)

# Neptune
xNeptune=30.05*np.sin(2*np.pi*t*163.7)
yNeptune=30.05*np.cos(2*np.pi*t*163.7)

# Pluto
xPluto=39.48*np.sin(2*np.pi*t*247.9)
yPluto=39.48*np.cos(2*np.pi*t*249.7)
```
