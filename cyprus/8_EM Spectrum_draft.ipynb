```python
plt.figure(figsize=(16,2))

x = np.arange(0.0, 8., 0.01) 
y = np.sin(2*np.pi*x)
yy= np.sin(2*np.pi*(1+x)*x/2);

# plt.plot(x,np.sin(2*np.pi*x),'b-',label='line')
plt.plot(x,y,'g')
plt.plot(x,yy,'b')


plt.ylabel('amplitude')
plt.xlabel('time')
```
