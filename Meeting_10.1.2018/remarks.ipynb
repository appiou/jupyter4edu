## Matplotlib notebook

 - modify individual lines, points or arrays
 - supports interactive plots

```python
%matplotlib notebook
from  ipywidgets import interact
import ipywidgets
import matplotlib.pyplot as plt
import numpy as np
```

```python
f, ax = plt.subplots(1,1,figsize=(5,5))
a,b,c = .2,.2,.2
c = np.sqrt(a**2-b**2)

t = np.linspace(0, 2*np.pi, 100)

ax.plot( a*np.cos(t) , b*np.sin(t) )
ax.plot([c,-c],[0,0],'o')
l1,l2 = ax.lines
#plt.axis('equal')
#plt.grid(color='lightgray',linestyle='--')
@interact(a=ipywidgets.widgets.FloatSlider(min=0,max=1,value=0.2),\
          b=ipywidgets.widgets.FloatSlider(min=0,max=1,value=0.1))
def _(a,b):
    l1.set_data(a*np.cos(t) , b*np.sin(t))
    
```

## K3d plots

for interactive 3d 

```python
import k3d
```

```python
plot  = k3d.plot()
plt_point = k3d.points([[-c,c,0]])
plt_line = k3d.line(np.stack([a*np.cos(t) , b*np.sin(t),np.zeros_like(t)]).T)
```

```python
plot += plt_line
plot += plt_point
plot
```

```python
plt_point.point_size =0.04
```

```python
plot.camera_auto_fit = False
for t_ in t:
    plt_point.positions = np.stack([[-c,c,0],[a*np.cos(t_) , b*np.sin(t_),0]])
```
